```python
class Solution(object):
    def minimumSumSubarray(self, nums, l, r):
        """
        :type nums: List[int]
        :type l: int
        :type r: int
        :rtype: int
        """
        n = len(nums)
        min_positive_sum = float('inf')
        
        # Iterate through all possible starting points
        for i in range(n):
            curr_sum = 0
            # Try all possible lengths between l and r
            for j in range(i, min(i + r, n)):
                curr_sum += nums[j]
                # Check if current length is within [l,r] range
                if j - i + 1 >= l and j - i + 1 <= r:
                    # Update minimum positive sum if current sum is positive
                    if curr_sum > 0:
                        min_positive_sum = min(min_positive_sum, curr_sum)
        
        return min_positive_sum if min_positive_sum != float('inf') else -1
    

solution = Solution()
solution.minimumSumSubarray([3, -2, 1, 4], 2, 3)

```

```python
class Solution(object):
    def minSumSubarray(self, nums, l, r):
        current_sum = 0  # To track the sum of the sliding window
        left = 0  # Left pointer of the window
        min_sum = float('inf')  # Store the minimum sum
        
        for right in range(len(nums)):
            current_sum += nums[right]  # Expand the window
            
            # If window size exceeds r, shrink from the left
            while (right - left + 1) > r:
                current_sum -= nums[left]
                left += 1

            # Check if window size is within [l, r] and sum > 0
            if l <= (right - left + 1) <= r and current_sum > 0:
                min_sum = min(min_sum, current_sum)
        
        return min_sum if min_sum != float('inf') else -1  # Return -1 if no valid subarray exists

solution = Solution()
solution.minSumSubarray([3, -2, 1, 4], 2, 3)
```
