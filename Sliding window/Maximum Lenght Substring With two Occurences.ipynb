```python
from collections import Counter

class Solution(object):
    def maximumLengthSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        n = len(s)
        count = Counter()
        left = 0
        longest_occurance = 0
        for right in range(n):
            count[s[right]] +=1
            
            # only expand if each element has at least one appearence
            while count[s[right]] > 2:
                count[s[left]] -= 1
                left += 1

            longest_occurance = max(longest_occurance, right - left+1)

        return longest_occurance
            
solution = Solution()
print(solution.maximumLengthSubstring('bcbbbcba'))
                               
```
