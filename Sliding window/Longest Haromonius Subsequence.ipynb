```python
from collections import Counter

class Solution(object):
    def findLHS(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        
        
        count_nums = Counter(nums)
        max_length = 0

        for num in nums:
            # store each element
            # num_el = count_nums.items()
            if num + 1 in count_nums:
                max_length = max(max_length, count_nums[num] + count_nums[num + 1])

        return max_length
    
solution = Solution()
print(solution.findLHS([1,3,2,2,5,2,3,7]))
```
