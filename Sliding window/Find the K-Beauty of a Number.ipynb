```python
class Solution(object):
    def divisorSubstrings(self, num, k):
        """
        :type num: int
        :type k: int
        :rtype: int
        """
        # convert num into string
        nums = str(num)

        # left = 0
        count = 0
        # len_substr = []

        for i in range(len(nums)-k+1):
            # expand the window
            # current_div += nums[right]
            current_div = nums[i:i+k]
            div = int(current_div)

            if div != 0 and num % div == 0:
                count += 1

        return count             
    

solution = Solution()
print(solution.divisorSubstrings(240, 2))
```
