```python
class Solution(object):
    def minimumDifference(self, nums, k):

        if k == 1:
            return 1
        
        sorted(nums)
        
        minDiff = float('inf')
        
        for i in  range(len(nums) - 1):
            minDiff = min(minDiff, nums[i+k-1] - nums[i])


        return minDiff
```
