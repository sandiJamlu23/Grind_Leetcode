```python
class Solution(object):
    def decrypt(self, code, k):
        """
        :type code: List[int]
        :type k: int
        :rtype: List[int]
        """
        n = len(code)
        result = [0] * n  # Initialize result array

        if k == 0:
            return result

        # Sliding window sum
        curr_sum = 0
        start, end = (1, k) if k > 0 else (n + k, n - 1)

        # Calculate initial sum
        for i in range(start, end + 1):
            curr_sum += code[i % n]

        # Populate result using sliding window
        for i in range(n):
            result[i] = curr_sum
            curr_sum -= code[(start) % n]  # Remove leftmost element
            start += 1
            end += 1
            curr_sum += code[end % n]  # Add new rightmost element

        return result

# Example usage
solution = Solution()
print(solution.decrypt([5, 7, 1, 4], 3))
```
