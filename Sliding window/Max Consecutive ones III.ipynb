```python
class Solution(object):
    def longestOnes(self,nums, k):

        left = 0
        zeros = 0
        max_cons = 0
        # right = 0

        for right in range(len(nums)):            
            if nums[right] == 0:
                zeros += 1

            # Shrink window if we have more zeros than allowed
            while zeros > k:
                if nums[left] == 0:
                    zeros -= 1
                left += 1

            max_cons = max(max_cons, right - left + 1)

        return max_cons

solution = Solution()
print(solution.longestOnes([0,0,0], 3))
```
