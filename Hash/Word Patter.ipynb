```python
# from collections import Counter
class Solution():
    def wordPattern(self,pattern, s):
        # count = Counter()

        # we need to split the pattern and s into arrays
        letter = s.split()

        pattern_to_s = {}
        s_to_pattern = {}

        if len(pattern) != len(letter):
            return False

        for i in range(len(pattern)):
            p = pattern[i]
            w = letter[i]
            # mapping
            if p in pattern_to_s:
                if pattern_to_s[p] != w:
                    return False
                
            elif w in s_to_pattern:
                if s_to_pattern[w] != p:
                    return False
            else:   
                pattern_to_s[p] = w
                s_to_pattern[w] = p
            # but how about the position?

        return True

solution = Solution()
solution.wordPattern("aaa","aa aa aa aa")

        
```
