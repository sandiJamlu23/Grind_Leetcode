```python
class Solution:
    def letterCombination(self, digits):
        """
        :type digits: str
        :rtype: List[str]
        """
        phone_num = {
            '1' : '',
            '2' : 'abc',
            '3' : 'def',
            '4' : 'ghi',
            '5' : 'jkl',
            '6' : 'mno',
            '7' : 'pqrs',
            '8' : 'tuv',
            '9' : 'wxyz'
        }

        # Base case: when we reach the end of the digits
        def backtrack(index, val):
            # If the length of val matches the digits length, we found a valid combination
            if index == len(digits):
                result.append(val)
                return

            # Get letters for the current digit
            digit = digits[index]
            letters = phone_num[digit]

            # For each letter in the current digit's corresponding string, recurse
            for letter in letters:
                backtrack(index + 1, val + letter)

        # If digits are empty, there's nothing to process
        if not digits:
            return []

        result = []
        backtrack(0, "")
        return result

# Testing
sol = Solution()
print(sol.letterCombination("23"))
print(sol.letterCombination("2"))
```
