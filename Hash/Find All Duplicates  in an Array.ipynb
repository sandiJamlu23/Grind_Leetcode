```python
from collections import Counter

class Solution():
    def findDuplicates(self, nums):
        count = Counter(nums)
        result = {}
        for key, value in count.items():
            if value > 1:
                result[key] = value
                
        return result.keys()
    

solution = Solution()
print(solution.findDuplicates([4,3,2,7,8,2,3,1]))
```

```python
class Solution():
    def findDuplicates(self, nums):
        result = []
        for num in nums:
            index = abs(num) - 1
            if nums[index] < 0:
                result.append(abs(num))
            else:
                nums[index] *= -1
        return result
    
solution = Solution()
print(solution.findDuplicates([4,3,2,7,8,2,3,1]))
```
