```python
class Solution():
    def setZeroes(self, matrix):

        m = len(matrix)
        n = len(matrix[0])
        
        row_to_zero = set()
        col_to_zero = set()

        for i in range(m):
            for j in range(n):
                if matrix[i][j] == 0:
                    row_to_zero.add(i)
                    col_to_zero.add(j)

        for i in range(m):
            for j in range(n):
                if i in row_to_zero or j in col_to_zero:
                    matrix[i][j] = 0

        return matrix
    

solution = Solution()
print(solution.setZeroes([[1,1,1],[1,0,1],[1,1,1]]))
```
