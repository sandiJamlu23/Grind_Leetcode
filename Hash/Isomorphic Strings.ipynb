```python
class Solution():
    def isIsomorphic(self, s, t):

        if len(s) != len(t):
            return False
        
        s_to_t = {}
        t_to_s = {}

        for i in range(len(t)):
            s_letter = t[i]
            t_letter = s[i]

            if s_letter in s_to_t:
                if s_to_t[s_letter] != t_letter:
                    return False
                
            else: s_to_t[s_letter] = t_letter


            if t_letter in t_to_s:
                if t_to_s[t_letter] != s_letter:
                    return False
                
            else: t_to_s[t_letter] = s_letter

        return True
    
solution = Solution()
solution.isIsomorphic('ad', 'eg')
```

# Or

```python
class Solution:
    def isIsomorphic(self, s, t):
        if len(s) != len(t):
            return False

        s_to_t = {}
        t_to_s = {}

        for sc, tc in zip(s, t):
            if sc in s_to_t:
                if s_to_t[sc] != tc:
                    return False
            else:
                s_to_t[sc] = tc

            if tc in t_to_s:
                if t_to_s[tc] != sc:
                    return False
            else:
                t_to_s[tc] = sc

        return True
```
