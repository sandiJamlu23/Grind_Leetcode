```python
from collections import Counter
class Solution():
    def majorityElement(self, nums):
        max_key = 0
        max_val = float('-inf')
        counter = Counter(nums)
        for key,val in counter.items():
            if val > max_val:
                max_val = val
                max_key = key

        return max_key  

solution = Solution()
print(solution.majorityElement([1,1,1,2,2]))      
```

 # Boyer-Moore Voting Algorithm

```python
class Solution:
    def majorityElement(self, nums):
        count = 0
        candidate = None

        for num in nums:
            if count == 0:
                candidate = num
            count += 1 if num == candidate else -1

        return candidate

solution = Solution()
print(solution.majorityElement([1,1,1,2,2]))
```
