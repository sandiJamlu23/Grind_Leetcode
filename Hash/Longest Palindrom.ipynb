```python
from collections import Counter
class Solution():
    def longestPalindrome(self, s):
        odd = False
        if len(s) == 1:
            return 1
        counter = Counter(s)
        total_sum = 0

        for key,values in counter.items():
            if values %2 == 0:
                total_sum += values
            elif values%2 == 1:
                total_sum += values -1
                odd = True
        if odd:
            return total_sum + 1
        return total_sum
    
sol = Solution()
print(sol.longestPalindrome("ccc"))
```
