```python
from collections import Counter

class Solution():
    def intersect(selef, num1, num2):
        
        # num1.sort()
        # num2.sort()
        result = []
        # count = Counter(num1)
        count = Counter(num2)

        for num in num1:
            # if num in num2 and count[num] > 0:
            if count[num] > 0:
                result.append(num)
                count[num] -= 1

        return result
    
solution = Solution()
print(solution.intersect([1, 2, 2, 3], [2, 2, 4, 5]))
```
