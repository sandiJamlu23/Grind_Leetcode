```python
class TreeNode(object):
    def __init__(self, val = 0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution(object):
    def checkTree(self, root):
        def sumRoot(root):
            if not root:
                return 0
            
            if not root.left and not root.right:
                return root.val

            # go to the left side of root
            most_left = sumRoot(root.left)
            most_right = sumRoot(root.right)

            # if the sum of root.left + root.right == root.val, return True
            if root.left and root.right and most_left + most_right == root.val:
                return root.val
            return 0

        return sumRoot(root) == root.val


root = TreeNode(10)
root.left = TreeNode(1)
root.right = TreeNode(6)

solution = Solution()
print(solution.checkTree(root))  # This should print True
```
