```python
class Solution(object):
    def isHappy(self, n):
        # number = []
        def get_square_sum(n):
            sum_square = 0
            while n > 0:
                digit = n % 10
                sum_square += digit ** 2
                n //= 10
            return sum_square

        first_loop = n
        second_loop = get_square_sum(n)

        while first_loop != 1 and first_loop != second_loop:
            first_loop = get_square_sum(first_loop)
            second_loop = get_square_sum(get_square_sum(second_loop))

        return first_loop == 1

        
solution = Solution()
print(solution.isHappy(19))
```
