```python
class Solution(object):
    def missingNumber(self, nums):
        
        # shouldve = |
        shouldve = set(range(len(nums)+1))
        actual = set(nums)
        return (shouldve-actual).pop()

solution = Solution()
print(solution.missingNumber([0,1]))
```

```python
class Solution(object):
    def missingNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n = len(nums)
        expected_sum = n * (n + 1) // 2  # Sum of first n natural numbers
        actual_sum = sum(nums)  # Sum of elements in nums
        return expected_sum - actual_sum  # The missing number

solution = Solution()
print(solution.missingNumber([3,0,1]))  # Output: 2
```
