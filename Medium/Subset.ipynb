```python
class Solution(object):
    def subset(self, nums):
        def backtrack(start, current):
    
            result.append(current[:])
          
            for i in range(start, len(nums)):
                current.append(nums[i]) # try current node
                backtrack(i+1, current) # try combination
                current.pop() # remove current node, and try another node
                            
        result = []
        backtrack(0, [])
        return result

solution = Solution()
n = [1,2,3]
print(solution.subset(n))
```
