```python
class Solution(object):
    def findDisappearedNumbers(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        # num_range = {}
        result = []
        # seen = False
        for i in range(1, len(nums)+1):
            if i not in nums:
                result.append(i)

        return result

sol = Solution()
print(sol.findDisappearedNumbers([1,1]))
```

```python
from collections import Counter
class Solution(object):
    def topKFrequent(self, words, k):
        """
        :type words: List[str]
        :type k: int
        :rtype: List[str]
        """
        count = Counter(words)
        res = []
        sorted_ct = sorted(count.items(), key=lambda item:item[1])
        # for key, value in count.items():
        res.append(sorted_ct)

        return res[:k]
    
sol = Solution()
print(sol.topKFrequent(["i","love","leetcode","i","love","coding"], 2))
```
