```python
class Solution(object):
    def pivotArray(self, nums, pivot):
        # Create separate lists for numbers less than, equal to, and greater than pivot
        less = []
        equal = []
        greater = []
        
        # Partition numbers into three groups
        for num in nums:
            if num < pivot:
                less.append(num)
            elif num == pivot:
                equal.append(num)
            else:
                greater.append(num)
                
        # Combine the three lists in order
        return less + equal + greater

solution = Solution()
solution.pivotArray([9,12,5,10,14,3,10],10)


```
