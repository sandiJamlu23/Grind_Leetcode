```python
class Solution:
    def reverse(self, x):
        # Handle negative numbers
        negative = -1 if x < 0 else 1
        x = abs(x)  # Work with the absolute value for simplicity
        
        reversed_num = 0
        while x != 0:
            # Extract the last digit
            digit = x % 10
            reversed_num = reversed_num * 10 + digit
            x //= 10  # Remove the last digit from x
            
            # Check for overflow
            if reversed_num > 2**31 - 1:
                return 0
        
        # Restore the sign
        reversed_num *= negative
        
        # Check for final overflow after restoring the sign
        if reversed_num < -2**31 or reversed_num > 2**31 - 1:
            return 0
        
        return reversed_num

# Testing
solution = Solution()
print(solution.reverse(-123))  # Output: -321
print(solution.reverse(12))    # Output: 21
```
