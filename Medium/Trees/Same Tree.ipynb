```python
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution(object):
    def isSameTree(self, p, q):
        def traverse(root):
            if not root:
                return ["None"]
            
            # traverse(root.left)
            # traverse(root.right)
            return [root.val] + traverse(root.left) + traverse(root.right)

        p_traverse = traverse(p)
        q_traverse = traverse(q)
        return p_traverse == q_traverse
    
rootP = TreeNode(1)
rootP.left = TreeNode(2)
rootP.right = None

rootQ = TreeNode(1)
rootQ.left = None
rootQ.right = TreeNode(2)

solution = Solution()
solution.isSameTree(rootP, rootQ)
```
