```python
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution(object):
    def buildTree(self, preorder, inorder):
        if not preorder or not inorder:
            return None
     
        # first index in preorder is root
        root_val = preorder.pop(0)
        # Create the root from TreeNode
        root = TreeNode(root_val)
        root_in_inorder = None

        # find the root in inorder
        for i in range(len(inorder)):
            if root_val == inorder[i]:
                root_in_inorder = i
                break
            # find the left side of the root in inorder (slicing)
        
        left_inorder = []
        right_inorder = []
        if root_in_inorder is not None:
            left_inorder = inorder[:root_in_inorder]
            right_inorder = inorder[root_in_inorder+1:]

        # make the left subtree
        root.left = self.buildTree(preorder, left_inorder)
        # make the right subtree
        root.right = self.buildTree(preorder, right_inorder)

        return root

    def fuckYouLeetcode(self, root): 
        if not root:
            return []
        result = []
        queue = [root]
        while queue:
            level_size = len(queue)
            level = []
            for i in range(level_size):
                node = queue.pop(0)
                if node:
                    level.append(node.val)
                    queue.append(node.left)
                    queue.append(node.right)
                else:
                    level.append(None)
            result.extend(level)
        #Remove trailing nulls.
        while result and result[-1] is None:
            result.pop()
        return result

solution = Solution()
root = solution.buildTree([3,9,20,15,7], [9,3,15,20,7])
print(solution.fuckYouLeetcode(root))
```
