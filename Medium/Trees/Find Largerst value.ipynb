```python
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution(object):
    def largestValues(self, root):
        result = []

        def dfs(root, depth):  # current_max is now a list
            if not root:
                return

            if depth == len(result):
                result.append(root.val)
            else:
                result[depth] = max(result[depth], root.val)  # Append the value from the list

            dfs(root.left, depth + 1)
            dfs(root.right, depth + 1)

        dfs(root, 0)  # Pass current_max as a list
        return result
                    
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)

solution = Solution()
print(solution.largestValues(root))
```
