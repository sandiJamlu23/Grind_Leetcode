```python
# Here is an example of BFS
def BFS(root):
    queue = []
    queue.append(root)

    while queue:
        n = len(queue)
        for _ in range(n):
            node = queue.pop(0) # the root

            if node.left:
                queue.append(node.left)

            if node.right:
                queue.append(node.right)
```

```python
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def levelOrder(root):
    if not root:
        return []
    
    queue = deque([root])
    result = []
    
    while queue:
        level = []
        '''In Python, _ is just a throwaway variable. It means:
            👉 "I don’t care about this value, I just need to loop a certain number of times."'''
        for _ in range(len(queue)):  # Process all nodes at the current level
            node = queue.popleft()
            ''' popleft() is a method from collections.deque, 
                and it removes and returns the leftmost element of the deque (queue).'''
            level.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        result.append(level)
    
    return result

# Quick win: Testing the function
root = TreeNode(3, TreeNode(9), TreeNode(20, TreeNode(15), TreeNode(7)))
print(levelOrder(root))  # Expected Output: [[3], [9, 20], [15, 7]]
```
