```python
class Solution(object):
    def permute(self, nums):                
        # to find all possible combination, we need to use all the possible combination
        # from each element
        def backtrack(used, current):
            # add the copy of current element to the result
            # base case
            if len(nums) == len(current):
                result.append(current[:])
                return # nothing
            
            for i in range(len(nums)):
                if i in used:
                    continue
                # choose the first element
                current.append(nums[i])
                used.add(i)
                backtrack(used, current)
                # pop the current element and try another one
                current.pop()
                used.remove(i)

        # store the result in an empty array
        result = []
        # nums.sort()
        backtrack(set(), [])
        return result


solution = Solution()
n = [1,2,3]
solution.permute(n)
```
