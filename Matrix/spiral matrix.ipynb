```python
class Solution:
    def spiralOrder(self, matrix):
        top = 0
        bottom = len(matrix) - 1
        left = 0
        right = len(matrix[0]) - 1

        res = []

        while left<= right and top<= bottom:
           
            # Go left to right across the top row
            for col in range(left, right+1):
                res.append(matrix[top][col])
            top +=1
                
            # Go top to bottom along the right column
            for row in range(top, bottom):
                res.append(matrix[row][right])
            right -=1

            # Go right to left across the bottom row (if top <= bottom)
            if top <= bottom:
                for col in range(right, left - 1, -1):
                    res.append(matrix[bottom][col])
                bottom -= 1

            # Go bottom to top along the left column (if left <= right)
            
            if left <= right:
                for row in range(bottom, top - 1, -1):
                    res.append(matrix[row][left])
                left += 1
        return res

sol = Solution()
print(sol.spiralOrder([[1,2,3],[4,5,6],[7,8,9]]))
```

```python
from collections import Counter
class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        k = 0
        count = Counter(nums)
        for key, val in count.items():
            
            while val > 2:
                val -=1
            if val <= 2:
                k += 1

        return k
sol = Solution()
print(sol.removeDuplicates(nums = [1,1,1,2,2,3]))

```

```python
class Solution(object):
    def removeDuplicates(self, nums):
        count = 0

        # remove the third duplicate

        for i in range(2, len(nums)):
            for j in range(i+1, len(nums)):
                if nums[i-2] == nums[j]:
                    pass
                elif nums[i-2] != nums[j]:
                    nums[j] = nums[i]
                    i +=1
        return count
        
sol = Solution()
print(sol.removeDuplicates(nums = [1,1,1,2,2,3]))
```

```python
class Solution(object):
    def removeDuplicates(self, nums):
        if len(nums) <= 2:
            return len(nums)
        
        i = 2  # write pointer

        for j in range(2, len(nums)):
            if nums[j] != nums[i - 2]:
                nums[i] = nums[j]
                i += 1
        
        return i
sol = Solution()
print(sol.removeDuplicates(nums = [1,1,1,2,2,3]))
```

```python
class Solution(object):
    def addStrings(self, num1, num2):
        """
        :type num1: str
        :type num2: str
        :rtype: str
        """
        i_n1=int(num1)
        i_n2=int(num2)
        return f"{i_n1 + i_n2}"

sol = Solution()
sol.addStrings(num1 = "456", num2 = "77")
        
```
