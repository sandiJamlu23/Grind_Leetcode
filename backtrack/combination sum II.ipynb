```python
# fix this
class Solution():
    def combinationSum2(self, candidates, target):
        def backtrack(index, current, comb_sum):
            if comb_sum == target:
                result.append(candidates[:])
                return 

            if comb_sum > target:
                return
            for i in range(index, len(candidates)):
                # check duplicates
                if i > index and candidates[i] == candidates[i-1]:
                    continue
                current.append(candidates[i])
                backtrack(i+1, current, candidates[i] + comb_sum)
                current.pop()

        candidates.sort()
        result = []
        backtrack(0, [], 0)
        return result
    
solution = Solution()
print(solution.combinationSum2([10,1,2,7,6,1,5], 8))
```
