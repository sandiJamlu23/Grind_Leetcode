```python
# brute force
class Solution(object):
    def countBits(self, n):
        ans = []
        for i in range(n+1):
            binary = bin(i)
            count = binary.count('1')
            ans.append(count)
        return ans


def count_ones(binary_str):
    count = 0
    for char in binary_str:
        if char == '1':
            count += 1
    return count

solution = Solution()
print(solution.countBits(5))

```

# Dynamic Programming

```python
class Solution(object):
    def countBits(self, n):
        result = [0] * (n+1)
        for i in range(1, n+1):
            result[i] = result[i//2] + (i%2)
        return result 
    
solution = Solution()
print(solution.countBits(5))
```
