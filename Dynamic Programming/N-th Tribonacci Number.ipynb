```python
class Solution(object):
    def tribonacci(self, n):
        if n == 0:
            return 0
            
        if n == 1 and n == 2:
            return 1
        
        dp = [0] * (n+1)
        dp[0] = 0
        dp[1] = 1
        dp[2] = 1
        
        # c = 2
        for i in range(3, n+1):
            dp[i] = dp[i-1] + dp[i-2] + dp[i-3]

        return dp[n]
    
solution = Solution()
solution.tribonacci(5)
```
