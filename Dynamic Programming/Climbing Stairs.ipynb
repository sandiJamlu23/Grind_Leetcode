```python
class Solution(object):
    def climbStairs(self, n):
        # total_comb = 0
        if n == 1:
            return 1
        if n == 2:
            return 2
        
        prev2 = 1
        prev1 = 2
       
        for i in range(2, n):
            current = prev2 +prev1
            prev2 = prev1
            prev1 = current

        return prev1
    

solution = Solution()
print(solution.climbStairs(3))
        
```
