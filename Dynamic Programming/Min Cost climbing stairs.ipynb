```python
class Solution(object):
    def minCostClimbingStairs(self, cost):


        n = len(cost)
        dp = [0] * n
        # min_cost = 0
        # start from costs at index 0
    
        for i in range(len(cost)):
            dp[i] = cost[i] + min(dp[i-1], dp[i-2])

        return min(dp[n-1],dp[n-2])
    

solution = Solution()
print(solution.minCostClimbingStairs([10,15,20]))
 
```

The O(1) space optimization comes from realizing that we dont actually need to store the entire dp[] array.

Why?
At any given step i, we only need the last two values:

dp[i-1]
dp[i-2]
So instead of using an array, we can just keep two variables and update them as we go.

Optimized Approach:
Start with prev2 = cost[0] and prev1 = cost[1] (the first two costs).
Iterate through the array, updating only two variables at a time.
The final answer will be the minimum of the last two computed values.

def minCostClimbingStairs(cost):
    prev2, prev1 = cost[0], cost[1]

    for i in range(2, len(cost)):
        curr = cost[i] + min(prev1, prev2)
        prev2, prev1 = prev1, curr  # Shift the values forward

    return min(prev1, prev2)  # Min cost to reach the top
This reduces the space complexity from O(n)  O(1) while keeping the time complexity at O(n).
