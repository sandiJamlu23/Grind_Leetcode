```python
class Solution(object):
    def countSquares(self, matrix):
        
        # the brute force way
        # loop through the row
        row = len(matrix)
        column = len(matrix[0])

        # make a dp matrix of 0's to populate later
        # that has the same size of input matrix
        dp = [[0]*column for _ in range(row)]
        total_squares = 0

        # go through all the row
        for i in range(row):
            # go through all the col
            for j in range(column):
                # find element 1
                if matrix[i][j] == 1:
                    # if the left or up bound of 1 is 0, return 1x1 matrix
                    if i == 0 or j == 0:
                        # populate the  first row or columne with 1
                        dp[i][j] = 1

                    else:
                        # check diagonally,left bound, and the upper bound
                        '''
                        The cell above (i-1, j)
                        The cell to the left (i, j-1)
                        The diagonal cell (i-1, j-1)
                        '''
                        dp[i][j] = min(dp[i-1][j], dp[i][j-1], dp[i-1][j-1])+1
                    total_squares += dp[i][j]

        return total_squares
        
solution = Solution()
print(solution.countSquares([
  [0,1,1,1],
  [1,1,1,1],
  [0,1,1,1]
]))
```
