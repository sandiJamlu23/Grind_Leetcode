```python
class Solution(object):
    def generate(self, num):
        # the total array is the same as the input i.e
        '''
        if input is 5, then we start from 1 and populate the empty array
        
        '''
        # we need to store the result in an empty array that keeps expanding until it reaches the same size as num
        # maybe initiate it with zeros (and then we ignore the zeros later) 
        dp = []

        # the first row
        # iterate until as long as the num
        for i in range(num):
            # so the i will consist of 1,2,3,4,5
            ''' base case'''
            row = [1] * (i+1) # first row is always 1
            for j in range(1, i):
                row[j] = dp[i-1][j-1] + dp[i-1][j] # add middle values
                '''
                dp[i-1][j-1]  The element diagonally left from the previous row.
                dp[i-1][j]  The element directly above from the previous row.
                
                '''
            dp.append(row)

        return dp

solution = Solution()
print(solution.generate(5))
```

Triangle pascal II

```python
class Solution(object):
    def getRow(self, rowIndex):
        dp = []
        for i in range(rowIndex + 1):
            row = [1] * (i+1)

            for j in range(1, i):
                row[j] = dp[i-1][j-1] + dp[i-1][j]
            dp.append(row)

        if rowIndex < len(dp):
            return dp[rowIndex]
        else:
            return []
        
solution = Solution()
print(solution.getRow(3))
```
