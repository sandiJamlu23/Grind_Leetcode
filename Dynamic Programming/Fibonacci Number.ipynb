```python
class Solution(object):
    def fib(self, n):
        # Base cases
        if n == 0:
            return 0
        if n == 1:
            return 1
        
        # Initialize first two numbers
        a, b = 0, 1
        
        # Calculate Fibonacci number
        for i in range(2, n + 1):
            a, b = b, a + b
            
        return b

solution = Solution()
print(solution.fib(2))  # Should print 1
```
