```python
class Solution:
    def isSubsequence(self, s, t):
        if not s:
            return True  # Empty string is always a subsequence
        if not t:
            return False  # Non-empty s cannot be a subsequence of an empty t
        
        left, right = -1, -1
        
        # Find first character of s in t
        for i in range(len(t)):
            if t[i] == s[0]:
                left = i
                break
        
        # Find last character of s in t
        for j in range(len(t) - 1, -1, -1):
            if t[j] == s[-1]:
                right = j
                break
        
        # Order must be preserved
        if left == -1 or right == -1 or left > right:
            return False
        
        # Now check if all middle characters of s exist in t[left:right] in correct order
        index = left  # Start from leftmost found character in t
        for k in range(1, len(s) - 1):  # Skip first and last character
            found = False
            while index < right:  # Search only in range (left, right)
                if t[index] == s[k]:
                    found = True
                    break
                index += 1
            if not found:
                return False
        
        return True

# Test cases
solution = Solution()
print(solution.isSubsequence("abc", "ahbgdc"))  # True
print(solution.isSubsequence("axc", "ahbgdc"))  # False
print(solution.isSubsequence("acb", "ahbgdc"))  # False
print(solution.isSubsequence("bca", "ahbgdc"))  # False
```
