```python
class Solution(object):
    def getLongestSubsequence(self, words, groups):
        # first_group = groups[0]
        prev_group = groups[0]
        subsequence = [words[0]]

        # while first_group < prev_group:
        for i in range(1, len(words)):
                if groups[i] != prev_group:
                    subsequence.append(words[i])
                    prev_group = groups[i]

        return subsequence

solution = Solution ()
solution.getLongestSubsequence(["e","a","b"], [0,0,1])
```
