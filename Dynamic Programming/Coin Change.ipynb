```python
class Solution(object):
    def change(self, amount, coins):
        dp = [0] * (amount+1)
        # result = []
        dp[0] = 1

        # for coin in coins:
        #     if amount in coin:
        #         return result.append(coin)
            
        for coin in coins:
            for i in range(coin, amount+1):
                dp[i] += dp[i-coin]

        return dp[amount]
    
solution = Solution()
print(solution.change(5,[1,2,5]))
```
