```python
class Solution(object):
    def generateParenthesis(self, n):
        result = []
        # if n == 1:
        #     result = ['()']

        # if n == 2:
        #     result = ['()()', '(())']

        # if n == 3:
        #     result = ["((()))", "(()())", "(())()", "()(())", "()()()"]

        def bactrack(current, openCount, closeCount):

            # base case
            if len(current) == 2*n:
                result.append(current)
                return

            if openCount < n:
                bactrack(current + '(', openCount + 1, closeCount)


            if closeCount < openCount:
                bactrack(current + ')', openCount, closeCount + 1)


        bactrack("", 0,0)

        return result
    

solution = Solution()
print(solution.generateParenthesis(3))
```
