```python
class Solution(object):
    def maxAbsoluteSum(self, nums):
        current_max, current_min = 0, 0
        max_abs, min_abs = 0, 0
        
        for num in nums:
            current_max = max(num, (current_max + num))
            current_min = min(num, (current_min + num))

            # find the absolute
            max_abs = max(current_max,max_abs)
            min_abs = min(current_min,min_abs)

        return max(max_abs, abs(min_abs))
    

solution = Solution()
print(solution.maxAbsoluteSum([1,-3,2,3,-4]))
```
