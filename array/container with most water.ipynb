```python
class Solution:
    def maxArea(self, height):
        left = 0
        right = len(height)-1
        max_water = 0

        while left < right:
            res = min(height[left], height[right]) * (right-left)
            max_water = max(max_water, res)

            # move the left pointer if the rght element is higher
            # if height[right] > height[left]:
            #     right -= 1
            if height[left] < height[right]:
                left += 1
            else: right -=1

        return max_water
    
sol = Solution()
print(sol.maxArea([1,8,6,2,5,4,8,3,7]))
```
