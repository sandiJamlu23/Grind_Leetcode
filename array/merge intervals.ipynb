```python
class Solution(object):
    def merge(self, intervals):
        """
        :type intervals: List[List[int]]
        :rtype: List[List[int]]
        """
        intervals.sort(key=lambda x:x[0])
        res = [intervals[0]]

        for current in intervals[1:]:
            if current[0] <= res[-1][1]:
                res[-1][1] = max(res[-1][1], current[1])

            else:
                res.append(current)

        return res
sol = Solution()
print(sol.merge([[1,3], [2,6], [8,10], [15,18]]))

                    
```

```python
class Solution():
    def reverse(self, x):
        # Handle negative numbers
        sign = -1 if x < 0 else 1
        x = abs(x)
        
        # Convert to string, reverse it, and convert back to integer
        reversed_x = sign * int(str(x)[::-1])
        
        # Check 32-bit integer range
        if reversed_x < -2**31 or reversed_x > 2**31-1:
            return 0
            
        return reversed_x
    
solution = Solution()
print(solution.reverse(-123))
```

```python
class Solution:
    def rotate(self, nums, k):
        n = len(nums)
        k = k % n  # just in case k > n
        
        while k > 0:
            temp = nums[-1]
            for i in range(n - 1, 0, -1):  # from n-1 down to 1
                nums[i] = nums[i - 1]
            nums[0] = temp
            k -= 1
        return nums

solution = Solution()
print(solution.rotate([1,2,3], 2))
```
