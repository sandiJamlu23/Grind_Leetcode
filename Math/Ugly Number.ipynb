```python
class Solution():
    def isUgly(self, n):

        # prime_num = [2,3,5]
        # even number aren't prime
        if n%2 == 0:
            return False

        # find the square root of the n
        for i in range(2,  int(n** 0.5)):
            if i%2 == 1:
                return False
            
        return True
            
sol = Solution()
print(sol.isUgly(6))
```

```python
class Solution:
    def isUgly(self, n):
        if n <= 0:
            return False
        
        # Divide the number by 2, 3, and 5 until it isn't divisible by them anymore
        for prime in [2, 3, 5]:
            while n % prime == 0:
                n //= prime
        
        # If the number becomes 1, it is an ugly number
        return n == 1

sol = Solution()
print(sol.isUgly(6))  # False (since 29 is not divisible by 2, 3, or 5)
```
