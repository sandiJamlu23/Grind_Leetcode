```python
class Solution(object):
    def findContentChildren(self, g, s):
        """
        :type g: List[int]
        :type s: List[int]
        :rtype: int
        """
        g.sort()
        s.sort()
        # result = []
        i,j = 0,0
        counter = 0

        while i < len(g) and j < len(s):
                if s[j] >= g[i]:
                    # dont compare each element to all elements 
                    # rather, compare each element to it's respective 
                    # position on the oposite list
                    i += 1
                j += 1
        return i
    
sol = Solution()
sol.findContentChildren( g = [1,2], s = [1,2,3])
```

```python
class Solution(object):
    def findContentChildren(self, g, s):
        i = 0  # children

        for j in range(len(s)):  # loop over cookies
            if i >= len(g):
                break  # no more children to satisfy

            if s[j] >= g[i]:
                i += 1  # satisfy child, move to next
        return i
    
sol = Solution()
sol.findContentChildren( g = [1,2], s = [1,2,3])
```
