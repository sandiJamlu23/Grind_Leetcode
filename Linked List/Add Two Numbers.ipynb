```python
class ListNode():
    def __init__(self, next=None, val=0):
        self.next = next
        self.val = val

class Solution(object):
    def addTwoNumbers(self, l1, l2):
        dummy = ListNode(0)
        current = dummy
        carry = 0
        while l1 or l2 or carry:
            v1 = l1.val if l1 else 0
            v2 = l2.val if l2 else 0

            # sum the val
            total = v1 + v2 + carry
            carry = total//10
            digit = total % 10

            current.next = ListNode(digit)
            current = current.next

            if l1:
                l1 = l1.next
            if l2:
                l2 = l2.next

        return dummy.next
    

solution = Solution()

root = ListNode()
```
